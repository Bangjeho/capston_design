```python
from sklearn.model_selection import GridSearchCV
import pandas as pd
import numpy as np
import keras
from keras.wrappers.scikit_learn import KerasRegressor
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.optimizers import Adam
```

```python
train_X = pd.read_csv('train_X.csv')
train_y = pd.read_csv('train_y.csv')
test_X = pd.read_csv('test_X.csv')
test_y = pd.read_csv('test_y.csv')

train_X = train_X.values
train_y = train_y.values
test_X = test_X.values
test_y = test_y.values
```

```python
def twolayer_build_model(nodesize):
    model = Sequential()
    print('input nodes : [%s][%s]'%(nodesize[0],nodesize[1]))
    model.add(Dense(nodesize[0],activation='relu'))
    model.add(Dense(nodesize[1],activation='relu'))
    model.add(Dense(1,activation='relu'))
    model.compile(loss='mse', optimizer='Adam', metrics=['mae'])
    return model

def onelayer_build_model(nodesize):
    model = Sequential()
    model.add(Dense(nodesize,activation ='relu'))
    model.add(Dense(1,activation = 'relu'))
    model.compile(loss = 'mse', optimizer='Adam', metrics = ['mae'])
    return model
```

```python
onelayer_model = KerasRegressor(onelayer_build_model, epochs = 25, batch_size = 100)
twolayer_model = KerasRegressor(twolayer_build_model, epochs = 10, batch_size = 100)
```
